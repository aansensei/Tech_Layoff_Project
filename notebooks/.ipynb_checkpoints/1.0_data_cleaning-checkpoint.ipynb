```python
import pandas as pd
import os

input_path = '../data/raw/tech_layoffs_raw.csv'
output_path = '../data/processed/tech_layoffs_clean.csv'

print(f"reading file from: {input_path}")
df = pd.read_csv(input_path)


df_clean = df.dropna(subset=['Laid_Off'])

df_clean = df_clean.rename(columns={
    'Company': 'company',
    'Location_HQ': 'location',
    'Industry': 'industry',
    'Laid_Off': 'laid_off_count',
    'Date_layoffs': 'date',
    'Stage': 'stage',
    'Country': 'country',
    'Money_Raised_in__mil': 'money_raised_mil',
    'Company_Size_before_Layoffs': 'company_size_before'
})

df_clean['date'] = pd.to_datetime(df_clean['date'])


os.makedirs('../data/processed', exist_ok=True)

df_clean.to_csv(output_path, index=False)

print(f"Done, cleaned data ({len(df_clean)} dòng) is saved at: {output_path}")
print(df_clean.head())
```
