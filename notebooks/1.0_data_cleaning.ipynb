```python
import pandas as pd
import os

current_dir = os.path.dirname(os.path.abspath(__file__)) if '__file__' in locals() else os.getcwd()

possible_paths = [
    '../data/raw/tech_layoffs_raw.csv',
    
]

valid_path = None
for path in possible_paths:
    full_path = os.path.normpath(os.path.join(current_dir, path))
    if os.path.exists(full_path):
        print(full_path)
        valid_path = full_path
        break

if valid_path:
    df = pd.read_csv(valid_path)
    
    df_clean = df.dropna(subset=['Laid_Off'])

    df_clean = df_clean.rename(columns={
        'Company': 'company',
        'Location_HQ': 'location',
        'Industry': 'industry',
        'Laid_Off': 'laid_off_count',
        'Date_layoffs': 'date',
        'Stage': 'stage',
        'Country': 'country',
        'Money_Raised_in__mil': 'money_raised_mil',
        'Company_Size_before_Layoffs': 'company_size_before'
    })

    df_clean['date'] = pd.to_datetime(df_clean['date'])

    try:
        output_dir = os.path.join(os.path.dirname(os.path.dirname(valid_path)), 'processed')
    except:
        output_dir = os.path.join(current_dir, 'data', 'processed')

    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, 'tech_layoffs_clean.csv')
```
